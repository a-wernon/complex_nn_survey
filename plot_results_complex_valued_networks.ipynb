```python
import matplotlib.pyplot as plt
import pickle
```

```python
# Load the pickled list back
pickles = [
    'first5',
    'next8',
    'tanh',
    'cardio',
    'tanh_and_relu_real'
]

results = []

for name in pickles:
    with open(f'{name}.pkl', 'rb') as file:
        loaded_list = pickle.load(file)
        results += loaded_list

# train_losses, test_losses, train_accuracies, test_accuracies, detailed_losses
```

```python
posib_names = [
    'SARelu',
    'SASigmoid',
    'SATanh',
    'SALeakyReLU',
    'SAELU',
    'ZDivideAbsZ',
    'SA_AmpPhRelu',
    'SA_AmpPhSigmoid',
    'SA_AmpPhTanh',
    'SA_AmpPhLeakyrelu',
    'SA_AmpPhElu',
    'ReLUOnMagnitude',
    'Sigmoid',
# aaa
    'Tahn',
    'Cardio',
    'Tahn_real',
    'ReLu_real'
]
```

# plot all train accuracy

```python

plt.figure(figsize=(20,10))
# train_losses, test_losses, train_accuracies, test_accuracies, detailed_losses
for i in range(len(results)):
    plt.plot(results[i][2], label=posib_names[i], linewidth=4)

#plt.yscale('log')
plt.legend()

plt.xlabel('Number of Iterations')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Number of Iterations')
```
